```python
import pandas as pd
import re

RawDat0001 = pd.read_csv('./data/savedrecs0001-1000.txt', sep="\t",index_col=False)
RawDat1000 = pd.read_csv('./data/savedrecs1001-2000.txt', sep="\t",index_col=False)
RawDat2000 = pd.read_csv('./data/savedrecs2001-3000.txt', sep="\t",index_col=False)
RawDat3000 = pd.read_csv('./data/savedrecs3001-4000.txt', sep="\t",index_col=False)
RawDat4000 = pd.read_csv('./data/savedrecs4001-5000.txt', sep="\t",index_col=False)
RawDat5000 = pd.read_csv('./data/savedrecs5001-6000.txt', sep="\t",index_col=False)
RawDat6000 = pd.read_csv('./data/savedrecs6001-7000.txt', sep="\t",index_col=False)
RawDat7000 = pd.read_csv('./data/savedrecs7001-7506.txt', sep="\t",index_col=False)

RawDat1000.index = RawDat1000.index + 1000
RawDat2000.index = RawDat2000.index + 2000
RawDat3000.index = RawDat3000.index + 3000
RawDat4000.index = RawDat4000.index + 4000
RawDat5000.index = RawDat5000.index + 5000
RawDat6000.index = RawDat6000.index + 6000
RawDat7000.index = RawDat7000.index + 7000

rawData = RawDat0001.append(
    RawDat1000.append(
    RawDat2000.append(
    RawDat3000.append(
    RawDat4000.append(
    RawDat5000.append(
    RawDat6000.append(
    RawDat7000)))))))
```

```python
## LDA analysis for clustering

# restrict to abstracts

abstracts = rawData[['AU','TI','AB']]

abstracts = abstracts.dropna()

# remove punctuation
abstracts['AB_processed'] = abstracts['AB'].map(lambda x: re.sub('[,\.!?]', '', x))
abstracts['AB_processed'] = abstracts['AB_processed'].map(lambda x: x.lower())

import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]

data = abstracts.AB_processed.values.tolist()
data_words = list(sent_to_words(data))
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])

import gensim.corpora as corpora
# Create Dictionary
id2word = corpora.Dictionary(data_words)
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
print(corpus[:1][0][:30])


from pprint import pprint
# number of topics
num_topics = 15
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
import pyLDAvis.gensim_models
import pickle 
import pyLDAvis
import os
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join('./results/ldavis_prepared_'+str(num_topics))
print(LDAvis_data_filepath)
```

```python
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
```

```python
from datetime import datetime
now = datetime.now()
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, './results/ldavis_prepared_'+ str(num_topics)+ now.strftime('%Y%m%d%H%M') +'.html')
LDAvis_prepared
```
